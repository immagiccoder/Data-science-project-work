# Step 3: Data Preparation

```python
# changing to 0 and 1 or encoding
data['fueltype'] = data['fueltype'].map({'gas' : 1 ,'diesel' : 0})
data['enginelocation'] = data['enginelocation'].map({'front' : 1, 'rear' : 0})
data['doornumber'] = data['doornumber'].map({'two' : 1 , 'four' : 0})
data['aspiration']  = data['aspiration'].map({'std' : 1,'turbo' : 0})
```

```python
# Creating Segment for Carname
```

```python
data.groupby('CarName').price.mean().sort_values(ascending=False)
```

```python
car_seg={
'jaguar':'High',        
'buick':'High',         
'porsche':'High',       
'bmw':'High',           
'volvo':'Medium',         
'audi':'Medium',          
'mercury':'Medium',       
'alfa':'Medium',          
'peugeot':'Medium',       
'saab':'Medium',          
'mazda':'Medium',         
'Nissan':'Medium',        
'volkswagen':'Medium',    
'toyota':'Low',         
'renault':'Low',         
'mitsubishi':'Low',      
'isuzu':'Low',           
'subaru':'Low',          
'honda':'Low',           
'plymouth':'Low',        
'dodge':'Low',           
'chevrolet':'Low'}
data['car_segment']=data['CarName'].map(car_seg)

data.head()
```

```python
# Drop CarName and CarID
data.drop('CarName',inplace=True,axis=1)
data.drop('car_ID',inplace=True,axis=1)
```

### Dummy Variables

```python
# Get Dummy Variables 

dummy_var = pd.get_dummies(data, columns = ['car_segment','cylindernumber','enginetype','fuelsystem','carbody','drivewheel'],drop_first = True)
```

```python
dummy_var.info()
```

```python
data.head()
```

# Step 4: Splitting the Data into Training and Testing Sets


```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
df_train, df_test = train_test_split(dummy_var, train_size = 0.7, test_size = 0.3, random_state = 100)
```

# Minmax Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
df_train.info()
```

```python
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables
num_vars = ['curbweight','carheight','boreratio','stroke','compressionratio','peakrpm','horsepower','citympg','price']


df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
df_train.head()
```

### Dividing into X and Y sets for the model building

```python
y_train = df_train.pop('price')
X_train = df_train
```

# Building our model

#### the LinearRegression function from SciKit Learn for its compatibility with RFE

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 70
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 10)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
#getting col which supports by rfe and assigning name c_rfe
c_rfe = X_train.columns[rfe.support_]
c_rfe
```

```python
col = X_train.columns[rfe.support_]
col
```

# Step 5: Building a linear model

#### we need to explicitly fit a constant using sm.add_constant(X) because if we don't perform this step, statsmodels fits a regression line passing through the origin, by default.

```python
#making a new dataframe with rfe variables
X_train_rfe = X_train[c_rfe]
```

```python
#importing statsmodel
import statsmodels.api as sm  
X_train_lm = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()   # Running the linear model
```

```python
# Print a summary of the linear regression model obtained
print(lm.summary())
```

#### enginelocation  is insignificant in presence of other variables; can be dropped

```python
X_train_new = X_train_lm.drop(["enginelocation"], axis = 1)
X_train_new.columns
```

```python
#Calculating VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

# Making Predictions

```python
#Applying the scaling on the test sets
num_vars = ['curbweight','carheight','boreratio','stroke','compressionratio','peakrpm','horsepower','citympg','price']

df_test[num_vars] = scaler.transform(df_test[num_vars])
```

```python
# Dividing into X_test and y_test
y_test = df_test.pop('price')
X_test = df_test
```

```python
# Now let's use our model to make predictions.

# Creating X_test_new dataframe by dropping variables from X_test
X_test_new = X_test[X_train_new.columns]

# Adding a constant variable 
X_test_new = sm.add_constant(X_test_new)
```

```python
# Making predictions
y_pred = lm.predict(X_test_new)
```

### Looking at the RMSE

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
#Returns the mean squared error; we'll take a square root
np.sqrt(mean_squared_error(y_test, y_pred))
```

### Checking the R-squared on the test set

```python
r_squared = r2_score(y_test, y_pred)
r_squared
```
